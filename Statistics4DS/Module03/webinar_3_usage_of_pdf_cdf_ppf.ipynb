## Gambling exercise

Assume that the winning can be described by Gaussian distribution with mean 4.14 and std. dev. 5.44.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import scipy.stats as stats

# Initialize distibution (Normal distribution)
win_dist = stats.norm
```

```python
x = np.arange(-30, 30, 0.005)

# Specifiy probability density function
win_pdf = win_dist.pdf(x, loc=4.14, scale=5.44)
```

```python
import scipy.stats as stats

# Instanstiate a Normal distribution, this will be Normal distribution with centre at 0 and std.dev. 1:
norm_dist = stats.norm

'''
Central Region with 95% of data will have two tails with 2.5% data in each.
There is a probability of 95% to fall in that region,
Using cdf-function function we may define this region between right border with cdf of 97.5%
and left border with cdf of 2.5 %.
'''

# correspnonding z-scores can be found using ppf() method
right_Z_score = norm_dist.ppf(0.975)
left_Z_score = norm_dist.ppf(0.025)

print("Right borber of the central 95 percent region is %1.2f" %right_Z_score)
print("Left borber of the central 95 percent region is %1.2f" %left_Z_score)
```

```python
plt.plot(x, win_pdf)
```

```python
# using this distribution function, we may compute probabilities
# the probability to win 4.14 or less may be defined as
# cumulative distribution function cdf(4.14)
# recall, cdf(x) is defined as probability that Value of Random variable is between -inf and x
# or simply less than x
# Also, the probability is defined as area under pdf curve

win_dist.cdf(4.14, loc=4.14, scale=5.44)
```

```python
average = 4.14
sigma = 5.44

plt.plot(x, win_pdf)

plt.fill_between(x=np.arange(-30,4.14,0.01), 
                 y1= stats.norm.pdf(np.arange(-30,4.14,0.01),loc=average, scale = sigma),
                 facecolor='blue')

```

```python
# Likewise the probability to loose is cdf(0), or the total probabilty to have negative outcome

win_dist.cdf(0, loc=4.14, scale=5.44)
```

```python
plt.plot(x, win_pdf)

plt.fill_between(x=np.arange(-30,0,0.01), 
                 y1= stats.norm.pdf(np.arange(-30,0,0.01),loc=average, scale = sigma),
                 facecolor='red')

```

```python
# the probabilty to win at least 5 dollars is the probility to lend right of the x=5 on the pdf curve

Pr_5_or_more = 1 - win_dist.cdf(5, loc=average, scale=sigma)

Pr_5_or_more
```

```python
plt.plot(x, win_pdf)

plt.fill_between(x=np.arange(5, 30 ,0.01), 
                 y1= stats.norm.pdf(np.arange(5, 30,0.01),loc=average, scale = sigma),
                 facecolor='yellow')

```

```python
# Lastly the probability to win between 7 and 8 dollars is
# cdf(8) - cdf(7)

win_dist.cdf(8, loc=average, scale=sigma) - win_dist.cdf(7, loc=average, scale=sigma)
```

```python
# There is an inverse function to cdf() - Percent point percentile (ppf()) 
# ppf(x) returns value such that the probability to land between -inf and this value is x
# or ppf(x) defines a point which cuts x fraction of observation

# To find point which cuts 50% of wins
win_dist.ppf(0.5, loc=average, scale=sigma)
```

```python
# if we interested in win value which cuts 25% of highest win 
# we should find ppf of (1 - 0.25)

win_dist.ppf(1-0.25, loc=average, scale=sigma)
```

It means that win value 7.81 dollars defined region of 25% of highest wins or the probality to land to the right of 7.81 is 25%

```python
plt.plot(x, win_pdf)

plt.fill_between(x=np.arange(7.81,30,0.01), 
                 y1= stats.norm.pdf(np.arange(7.81,30,0.01),loc=average, scale = sigma),
                 facecolor='yellow')

plt.text(x=8, y=0.025, s="25% highest")
```
