# Probability density functions (PDFs) for a continuous random variable

In the case of a continuous random variable, the probability of any individual outcome is theoretically zero because a continuous random variable is one that can assume an uncountable or infinite number of values. As such, we cannot list the possible values because there is an infinite number of them!

However, we can determine the probability of a range of values. We can define a formula that can be used to compute the probabilities of a range of outcomes for a continuous variable. The graph of a continuous probability distribution is a curve, and probability is represented by the **area under the curve.**

The curve is called the **probability density function** (PDF), and we use the symbol $f(x)$ to represent it. $f(x)$ is the function that corresponds to the graph &mdash; we use it to graph the probability distribution.

The following requirements apply to a probability density function $f(x)$ whose range is $a ≤ x ≤ b$:

1. The domain of $f$ must be the set of all possible states of $x$<br><br>

2. $f(x) ≥ 0$ for all $x$<br><br>

3. The total area under the curve is 1<br><br>

4. The probability that $x$ has a value within the range $a ≤ x ≤ b$ is the area under the curve between $x=a$ and $x=b$

A function which describes the **area under the curve** is called a **cumulative distribution function** (CDF). The cumulative distribution function is used to evaluate probability as area. CDFs have the following properties:

1. The outcomes are measured, not counted<br><br>

2. The entire area under the curve and above the $x$-axis is equal to 1<br><br>

3. Probability is found for intervals (ranges) of $x$ values rather than for individual $x$ values<br><br>

4. $P(a < x < b)$ is the probability that the random variable $X$ is in the interval between the values $a$ and $b$. $P(a < x < b)$ is the area under the curve, above the $x$-axis, to the right of $a$ and the left of $b$.<br><br>

5. $P(x = a) = 0$ The probability that $x$ takes on any single individual value is zero. The area below the curve, above the $x$-axis, and between $x = a$ and $x = a$ has no width, and therefore no area (area = 0). Since the probability is equal to the area, the probability is also zero.<br><br>

6. $P(a < x < b)$ is the same as $P(a ≤ x ≤ b)$ since probability is equal to area

## Uniform distribution (for continuous variables)

There is a version of the uniform distribution for continuous variables.  The continuous uniform distribution is a probability distribution and is concerned with events that are equally likely to occur. When working out problems that have a uniform distribution, be careful to confirm if the data is inclusive or exclusive of endpoints.

Examples of continuous uniform distributions include:

* The angle a spinner will land on
* The time of arrival of the next bus (if you don't have a schedule)
* Rounding error (i.e. the difference between a rounded number and the original number)

The uniform distribution is characterized as:

$$
f(x) =     \left\{ \begin{array}{rcl}
         {\frac{1}{b-a}} & \mbox{for}
         & a \leqq x \leqq b \\ 0  & \mbox{} & \textrm{otherwise} \\
         \end{array}\right.
$$

The mean, $\mu$, and variance, $\sigma^{2}$, for the uniform distribution are:

 $$
 \mu=\frac{a+b}{2}
 $$
 
 $$
 \sigma^{2}=\frac{(b-a)^{2}}{12}
 $$

The notation for a uniform distribution is $X \sim U(a,b)$ where $a$ is the lowest value of $x$ and $b$ is the highest value of $x$.

**Example**

Let's simulate a uniform distribution for a continuous variable.

```python
import numpy as np
nobs = 100
min_val = 1
max_val = 100
Y = np.random.uniform(min_val, max_val, nobs)
Y
```

## Normal distribution

Among all the distributions we see in practice, the **normal distribution** is overwhelmingly the most common. The symmetric, unimodal, bell curve is ubiquitous throughout statistics. Indeed it is so common, that people often know it as the normal curve or normal distribution.

You see the bell curve is almost everywhere around us. For example: the height of trees and people, grades of students in a course, income across members of a population, sugar level of patients, etc.

The first record of the normal distribution is in Abraham de Moivre's book <i>Doctrine of Chances</i> (De Moivre, 1718). De Moivre obtained the normal distribution as an approximation of the binomial distribution for a large number of trials. His results were praised by gamblers. He based his work on Newton’s infinitesimal calculus (Newton and de Moivre were friends).

The probability density function for the normal distribution is:

$$  f(x,\mu,\sigma) = \frac{1}{\sigma\sqrt{2\pi}}e^\frac{-(x-\mu)^2}{2\sigma^2} $$

Key properties of the normal distribution include: 

* It is defined by two parameters, $\mu$ and $\sigma$


* It is symmetric and centered around the mean, $\mu$


* It is unimodal


* It is bell-shaped

In the case of the normal distribution, the mean $\mu$ is the same as the median. Since the area under the curve must equal one, a change in the standard deviation, $\sigma$, causes a change in the shape of the curve: the curve becomes wider or narrower depending on $\sigma$. A change in $\mu$ causes the graph to shift to the left or right. Thus, there are an infinite number of normal probability distributions.

The notation for a normal distribution is  $X \sim N(\mu,\sigma)$, which is read as $X$ is a normal distribution with average $\mu$, and standard deviation $\sigma$.

Of special interest is the **standard normal distribution**, $Z \sim N(0,1)$
where $Z$, called the $Z$-score, for a normal distribution $X \sim N(\mu,\sigma)$ is defined as

$$
Z=\frac{x-\mu}{\sigma}
$$

If $X \sim N(\mu,\sigma)$ then $Z=\frac{x-\mu}{\sigma} \sim N(0,1)$

The $Z$-score of an observation is defined as the number of standard deviations it falls above or below the mean. If the observation is one standard deviation above the mean, then its $Z$-score is 1. If the observation is 1.5 standard deviations below the mean, then its $Z$-score is -1.5.

Any normal distribution can be transformed into a standard normal distribution using the equation given above.  

We can use $Z$-scores to roughly identify which observations are more unusual than others. One observation $x_1$ is said to be more unusual than another observation $x_2$ if the absolute value of its $Z$-score is larger than the absolute value of the other observation’s $Z$-score: $\mid Z_1 \mid > \mid Z_2 \mid$. This technique is especially insightful when a distribution is symmetric, as in the case of the normal distribution.

**The empirical rule**

If $X$ is a random variable and has a normal distribution with mean $\mu$ and standard deviation $\sigma$, then the empirical rule states the following:

* About 68% of the $x$ values lie between –1$\sigma$ and +1$\sigma$ of the mean $\mu$ (within one standard deviation of the mean)


* About 95% of the $x$ values lie between –2$\sigma$ and +2$\sigma$ of the mean $\mu$ (within two standard deviations of the mean)


* About 99.7% of the $x$ values lie between –3$\sigma$ and +3$\sigma$ of the mean $\mu$ (within three standard deviations of the mean). Notice that almost all the $x$ values lie within three standard deviations of the mean.


* The $Z$-scores for +1$\sigma$ and –1$\sigma$ are +1 and –1, respectively


* The $Z$-scores for +2$\sigma$ and –2$\sigma$ are +2 and –2, respectively


* The $Z$-scores for +3$\sigma$ and –3$\sigma$ are +3 and –3 respectively


The empirical rule is also known as the **68-95-99.7** rule.

The shaded area in the following graph indicates the area to the left of $x$. This area is represented by the probability $P(X < x)$. Normal distribution tables, computers, and calculators provide or calculate the probability $P(X < x)$.


* The area to the right is then $P(X > x) = 1 – P(X < x)$. 


* $P(X < x)$ = Area to the left of the vertical line through $x$


* $P(X > x) = 1 – P(X < x)$ = Area to the right of the vertical line through $x$


* $P(X < x)$ is the same as $P(X ≤ x)$, and $P(X > x)$ is the same as $P(X ≥ x)$ for continuous distributions

### Example of probability calculation using the normal distribution

```python
# Simulate a data set of 10000
mu = 100.0
sigma = 25.0
normal_data = np.random.normal(mu, sigma, 10000)

# Show first 10 samples
normal_data[:10]
```

Let's suppose we want to calculate the probability of a value of 75 or less in the normal distribution we just simulated. 

```python
# First take a look at the pdf and especially the dark blue area under
# the curve containing the probability of 75 or less.
x = 75.0
sns.distplot(normal_data, label='Data', kde=False, 
             fit=stats.norm, color='b')
plt.text(x+5, .0003, '$x$='+str(x))

x_plot = np.linspace(min(normal_data), x, 1000)
y_plot = stats.norm.pdf(x_plot, mu, sigma)
plt.fill_between(x_plot,  y_plot)
c=plt.legend()
```

```python
# To calculate the probability, we need the Z-score.
zscore = (x - mu) / sigma
zscore
```

```python
# Calculate the probability by calling stats.norm.cdf.  Note that the value matches the empirical rule.
p = stats.norm.cdf(zscore)
p
```

We can validate the empirical rule using simulation.

```python
# Create a boolean vector (like 0 and 1), which evaluates a condition. sum() adds the ones => it is like counting
(abs(normal_data-normal_data.mean())<=normal_data.std())

# Count the number of data points between the average +/- one standard deviation
(abs(normal_data-normal_data.mean())<=normal_data.std()).sum()

# Count the number of elements (= population_size)
normal_data.size

print('Number of data points between mean +/- one standard deviation:',
      round((abs(normal_data-normal_data.mean())<=normal_data.std()).sum()/normal_data.size*100),'%')
print('Number of data points between mean +/- 1.96 standard deviation:',
      round((abs(normal_data-normal_data.mean())<=1.96*normal_data.std()).sum()/normal_data.size*100),'%')
print('Number of data points between mean +/- 2.58 standard deviation:',
      round((abs(normal_data-normal_data.mean())<=2.58* normal_data.std()).sum()/normal_data.size*100,1),'%')
```

It is more common to apply the empirical rule to the standard normal distribution.  Notice how the standard normal distribution has a mean equal to zero and a standard deviation of 1.

```python
plt.hist((normal_data-normal_data.mean())/normal_data.std(),50)
plt.show()

print('For the standard normal distribution:')
print('Number of data points between +/- 1:',
      round((abs((normal_data-normal_data.mean())/normal_data.std())<=1).sum()/normal_data.size*100),'%')
print('Number of data points between +/- 1.96:',
      round((abs(normal_data-normal_data.mean())<=1.96*normal_data.std()).sum()/normal_data.size*100),'%')
print('Number of data points between +/- 2.58:',
      round((abs(normal_data-normal_data.mean())<=2.58* normal_data.std()).sum()/normal_data.size*100,1),'%')
```

# Central Limit Theorem

The Central Limit Theorem (CLT for short) is one of the most powerful and useful ideas in all of statistics. 

The CLT can be expressed in two alternative ways, and both alternatives are concerned with drawing finite samples of size $n$ from a population with a known mean, $\mu$, and a known standard deviation, $\sigma$. Because samples are random, the mean (or sum) of the sample will never be exactly the same. In fact, they follow a distribution usually called the distribution of sample means or of sample sums.

The power of the theorem is in the fact that it does not matter much what the shape of the distribution of the original population is: the sample means (or sums) are going to follow a normal distribution as long as the samples are reasonably independent and of similar variance.

* The first definition says that if we collect samples of size $n$ with a "large enough $n$," calculate each sample's mean, and create a histogram of those means, then the resulting histogram will tend to have an approximate normal bell shape with mean equal $\mu$ and standard deviation $\sigma_x = \frac{\sigma}{\sqrt{n}}$. 
* The second definition says that if we again collect samples of size $n$ that are "large enough," calculate the sum of each sample, and create a histogram, then the resulting histogram will again tend to have a normal bell shape with $\mu_x= n \times \mu$ and with standard deviation $\sigma_x^{2}= n \times \sigma^{2}$.

The size of the sample, $n$, that is required in order to be "large enough" depends on the original population from which the samples are drawn (as a general rule of thumb, the sample size should be at least 30 or the data should come from a normal distribution). If the original population is far from normal, then more observations are needed for the sample means or sums to be normal. Sampling is done with replacement.

It would be difficult to overstate the importance of the central limit theorem in statistical theory. Knowing that the sample means from any data source behave in a predictable and known way &mdash; even if the original distribution is not normal &mdash; is a powerful tool.

The CLT also explains why the normal distribution appears so often in nature. Any time a quantity we measure is the additive result of many independent physical processes (as is often the case) the quantity will manifest a normal-like distribution.

## The Central Limit Theorem for sample means (averages)

The CLT is the foundation of inference, hypothesis testing, regression, and many other fields in classical statistics.  Even some computational techniques apply this theorem.

Suppose $X$ is a random variable with a distribution that may be known or unknown (it can be any distribution). Using a subscript that matches the random variable, suppose:

1. $\mu_X$ = the mean of $X$<br><br>

2. $\sigma_X$ = the standard deviation of $X$

If you draw random samples of size $n$, then as $n$ increases, the random variable $\bar X$, which consists of sample means, tends to be normally distributed, and 
$\bar X\sim N(\mu,\frac{\sigma_X}{\sqrt{n}})$.

The Central Limit Theorem for sample means says that if you keep drawing larger and larger samples (such as rolling one, two, five, and finally, ten dice) and calculating their means, the sample means form their own normal distribution (the sampling distribution). The normal distribution has the same mean as the original distribution and a variance that equals the original variance divided by the sample size. 

To put it more formally, if you draw random samples of size $n$, the distribution of the random variable $\bar X$, which consists of sample means, is called the sampling distribution of the mean. The sampling distribution of the mean approaches a normal distribution as $n$, the sample size, increases.

The random variable $\bar X$ has a different $Z$-score associated with it from that of the random variable $X$. The mean $\bar X$ is the value of $\bar X$ in one sample.
 $Z = \frac{\bar X −\mu_x}{\frac{\sigma_X}{\sqrt{n}}}$
$\mu_X$ is the average of both $X$ and $\bar X$.
$\sigma_\bar X = \frac{\sigma_X}{\sqrt{n}}$ = standard deviation of $\bar X$ and is called the standard error of the mean.

## The Central Limit Theorem for sample sums

Suppose $X$ is a random variable with a distribution that may be known or unknown (it can be any distribution) and suppose:

1. ${\mu}_x$ = the mean of $X$<br><br>

2. ${\sigma}_x$ = the standard deviation of $X$

If you draw random samples of size $n$, then as $n$ increases, the random variable $\sum X$ consisting of sums tends to be normally distributed and $\sum X \sim N(n \mu_X,\sqrt{n}\sigma_X)$.

where:
* the mean is the mean of the original distribution
* the standard deviation is the standard deviation of the original distribution
* sample size = $n$

The Central Limit Theorem for sums says that if you keep drawing larger and larger samples and taking their sums, the sums form their own normal distribution (the sampling distribution), which approaches a normal distribution as the sample size increases. The normal distribution has a mean equal to the original mean multiplied by the sample size and a standard deviation equal to the original standard deviation multiplied by the square root of the sample size.

## Using the Central Limit Theorem

It is important for you to understand when to use the Central Limit Theorem. If you are being asked to find the probability of the mean, use the CLT for the mean. If you are being asked to find the probability of a sum or total, use the CLT for sums. This also applies to percentiles for means and sums.

<blockquote><b>NOTE</b>: This is a very common mistake. If you are being asked to find the probability of an individual value, do not use the CLT. Use the distribution of its random variable.</blockquote>

Notice that the theorem only asks for data that is independent and identically distributed (iid), not for normality.  The big application of this is that the sampling distribution of the means of any distribution will follow a normal distribution.

Let's see an example starting from a uniform distribution:

1. Let's suppose we have a population that follows a uniform distribution
2. We take samples and we take the average of each sample
3. We analyze the distribution of the averages

**Example**

Let's simulate a uniform population from which we will take random samples and validate the CLT.

```python
population = np.random.uniform(0,100,1000000)
plt.hist(population)
print('population average:           ',population.mean())
print('population standard deviation:', population.std())
```

```python
nbr_samples = 20000 # rows
sample_size = 30 # columns
samples = np.random.choice(population,(nbr_samples,sample_size),replace = False)
samples
```

```python
means = samples.mean(1) # 1 because I want samples by row
plt.hist(means, bins = 50)
plt.show()
```

```python
means
```

```python
print('sample mean mean:   ',means.mean())
print('sample mean std:     ',means.std())
```

The example confirms that the mean of the population $\mu$ converges to the mean of the sample distribution and that standard deviation of the population and the sample distribution of the mean follow the following relation: $\sigma/\sqrt{n}$

```python
print('population average:           ',round(population.mean(),3), ' ~ ', 'sample mean mean:   ',
      round(means.mean(),3))
print('population standard deviation/square root (sample size) = ',round(population.std(),2), '/',
      round(np.sqrt(sample_size),2), '=',round(population.std()/np.sqrt(sample_size),2),'~ sample mean std=',
      round(means.std(),2))
```

A key word in our previous definition is "when the sample size increases." A sample size that is "big enough" to apply the CLT will depend on the shape of the population you are sampling from. See in the example below how the sample mean converges to the population mean as the sample size increases.

The code below calculates a rolling average to show convergence as the sample size increases.

```python
# Generate the population
mean_population = 50
stddev_population =10
population_size = 1000000
normal_data = np.random.normal(mean_population,stddev_population,population_size)

# Take samples from the population
nbr_samples = 5000 # rows
sample_size = 30 # columns
samples = np.random.choice(population,(nbr_samples,sample_size),replace = False)

# Generate an array that counts the number of samples taken
sample_number_array=np.arange(1,nbr_samples+1)

# Calculate the rolling average
rolling_average = samples.mean(1).cumsum()/sample_number_array
rolling_average

# Plot the rolling average
plt.scatter(sample_number_array,rolling_average,alpha=0.5,marker=None,linewidths=1)
plt.show()
```

We can define a function to see the randomness of the process in one set of charts.

```python
def method():
    samples = np.random.choice(population,(nbr_samples,sample_size),replace = False)
    rolling_average = samples.mean(1).cumsum()/sample_number_array
    plt.scatter(sample_number_array,rolling_average,alpha=0.5,marker=None,linewidths=1)
    plt.ylabel("Simulated Frequency")
    plt.xlabel("Rolling Average")
    plt.show()

for i in range(5):
    method()
```

Another important concept that is very important in inference is that the sample means are also a random variable.  To illustrate this, let's plot the sample means just calculated.

```python
plt.figure(figsize=(16,6))
plt.scatter(sample_number_array,samples.mean(1),marker ='.',label = 'Sample Means from the same population')
plt.xlabel("Simulation Iteration")
plt.ylabel("Sample Average")
plt.legend(loc=2)
plt.show()
```
